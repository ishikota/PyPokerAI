### 15万iteratino後のweights同士で対戦

- あるプレイヤのhistoryが0.9(持ち金の9割をpayしてる)と大きいのを見てfoldを選べている
  - と思ったが,playerのhistoryに対する重みの絶対値が1を超えていて,大きすぎる気もする
- foldしまくって10round終えるパターンが多そう
  - 1round目の最初のplayがraiseするけど,次にallinされてfoldしてからは,あるプレイやのstackが大きくなるのでfoldしやすくなる?
    - foldしない時もちゃんとあった. (cardsが強いのが理由の上位にきてる)
- ALLINされるとPOTが大きすぎるのでFOLDが強くなる
